## Remove outlier

```python
from scipy import stats
# Function to remove outliers using Z-Score
def remove_outliers_zscore(df, zscore_threshold=3):
    z_scores = stats.zscore(df)
    abs_z_scores = abs(z_scores)
    filtered_entries = (abs_z_scores < zscore_threshold).all(axis=1)
    return df[filtered_entries]

# Removing outliers using Z-Score
df = remove_outliers_zscore(df[['Age', 'Tenure in Months', 'Avg Monthly Long Distance Charges', 'Avg Monthly GB Download', 'Total Revenue', 'Average Monthly Revenue', 'Total Additional Charges']])

print("Original DataFrame Shape:", df.shape)
```
